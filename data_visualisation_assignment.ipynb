```python
#Question4:Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
#          variable across different categories.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate a consistent dataset
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [np.random.normal(loc, scale, 100) for loc, scale in zip(range(10, 50, 10), range(5, 15, 3))]

# Ensure all arrays are of the same length
values = np.concatenate(values)

# Create corresponding categories
category_labels = np.repeat(categories, 100)

# Convert the dataset into a DataFrame
df = pd.DataFrame({'Category': category_labels, 'Values': values})
print("Dataset:")
print(df.head())

# Create a box plot to visualize the distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Values', data=df, palette='Set3')

# Add titles and labels
plt.title("Distribution of Values Across Different Categories")
plt.xlabel("Category")
plt.ylabel("Values")

# Show the plot
plt.show()

```

```python
#Question 5:Generate a synthetic dataset with correlated features.
            #Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset
size = 100  # Number of samples

# Generate correlated data
x1 = np.random.normal(0, 1, size)
x2 = 0.8 * x1 + np.random.normal(0, 0.1, size)
x3 = 0.6 * x1 + 0.2 * x2 + np.random.normal(0, 0.1, size)
x4 = np.random.normal(0, 1, size)

# Create a DataFrame
data = pd.DataFrame({'Feature1': x1, 'Feature2': x2, 'Feature3': x3, 'Feature4': x4})
print("Dataset:")
print(data.head())

# Compute the correlation matrix
corr_matrix = data.corr()
print("\nCorrelation Matrix:")
print(corr_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Question3:Using the sales data, generate a heatmap to visualize the variation in sales across different months and
#days.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the sales dataset
np.random.seed(20)
sales_data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
sales_df = pd.DataFrame(sales_data)

# Pivot the data to create a matrix of sales by month and day
sales_pivot = sales_df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum', fill_value=0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(sales_pivot, annot=True, cmap='YlGnBu', linewidths=0.5)

# Add titles and labels
plt.title("Heatmap of Sales Data Across Months and Days")
plt.xlabel("Month")
plt.ylabel("Day")

# Show the plot
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function
# z=sin(x^2+Y^2)^0.5

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis')

# Add a color bar which maps values to colors
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

# Add titles and labels
ax.set_title("3D Surface Plot of \( z = \sin(\sqrt{x^2 + y^2}) \)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Show the plot
plt.show()
```

```python
#Question:Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="y=sin(x)", line_width=2)

# Show the results
output_notebook()
show(p)
```

```python
#Question 2:2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#            markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np

# Prepare the data
np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.rand(100) * 40 + 10  # Marker sizes between 10 and 50
colors = np.random.choice(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], 100)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot
p = figure(title="Bokeh Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer with sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the results
output_notebook()
show(p)
```

```python
#Question 3: Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
data = {'fruits': fruits, 'counts': counts}

# Create a ColumnDataSource
source = ColumnDataSource(data=data)

# Create a new plot
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=['#f46d43', '#66c2a5', '#fee08b', '#3288bd'], factors=fruits))

# Customize plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the results
output_notebook()
show(p)
```

```python
#Question4:Create a Bokeh histogram to visualize the distribution of the given data
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph with the histogram data
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.5)

# Show the results
output_notebook()
show(p)
```

```python
#Question5:Create a Bokeh heatmap using the provided dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
import numpy as np

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource
data = {'x': x_flat, 'y': y_flat, 'z': z_flat}

from bokeh.models import ColumnDataSource
source = ColumnDataSource(data=data)

# Define the color mapper
mapper = LinearColorMapper(palette="Viridis256", low=np.min(z_flat), high=np.max(z_flat))

# Create a new plot
p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y', tooltips=[("x", "@x"), ("y", "@y"), ("value", "@z")])

# Add a quad glyph
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=transform('z', mapper), line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper, label_standoff=12, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the results
output_notebook()
show(p)
```
